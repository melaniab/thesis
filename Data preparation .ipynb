```python
b = b.sort_values(by=['goodreads_book_id'])
b.head()
```

```python
b_reduced = b[b["goodreads_book_id"] <=32085]
```

```python
len(b_reduced)
```

```python
b_new = b_reduced[['goodreads_book_id', 'authors', 'original_publication_year',
       'original_title', 'language_code']]
b_new.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# b_scaled = pd.DataFrame(scaler.fit_transform(b_new['original_publication_year']), columns=b_new.columns)
# b_scaled.head()
b_new['year_scaled'] = scaler.fit_transform(b_new['original_publication_year'].values.reshape(-1,1))
```

```python
b_new.tail()
```

```python
def map_to_en(x):
    if x in ['en', 'en-CA', 'en-GB', 'en-US', 'eng']: 
        return "en"
    else:
        return x
    
b_new["language_code"] = b_new["language_code"].apply(map_to_en)
b_new["language_code"].value_counts()
```

```python
# use pd.concat to join the new columns with your original dataframe
b_new = pd.concat([b_new,pd.get_dummies(b_new['language_code'], prefix='lang')],axis=1)

# now drop the original 'country' column (you don't need it anymore)
b_new.drop(['language_code'],axis=1, inplace=True)
```

```python
b_new.tail()
```

```python
b_new.tail()
```

```python
def string_to_list(string): 
    return string.split(", ")
b_new["authors_list"] = b_new["authors"].apply(string_to_list)
b_new.tail()
```

```python
vocab = np.array(b_new["authors_list"]).flatten()
```

```python
#flatten the 2D array 
authors_vocab =  [item for sublist in vocab for item in sublist]
```

```python
authors_vocab = list(set(authors_vocab))
```

```python
authors_vocab 
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

df_authors = pd.DataFrame(mlb.fit_transform(b_new["authors_list"]),columns=mlb.classes_, index=b_new["authors_list"].index)
```

```python
df_authors.tail()
```

```python
#TODO - title as TF-IDF? 
```

```python
b_feat = b_new[["goodreads_book_id", "year_scaled", "lang_en", "lang_spa", "lang_fre", "lang_ger", "lang_ita"]]
```

```python
b_feat.head()
```

```python
#TODO: dimentionality reduction na  df       
#TODO:     content-based  
```

```python
#Reduce dimentionality
from sklearn.decomposition import TruncatedSVD

arr_authors_reduced = TruncatedSVD(n_components=50, random_state=0).fit_transform(df_authors)
```

```python
df_authors_reduced = pd.DataFrame(arr_authors_reduced, index = b_feat.index)
```

```python
df_authors_reduced.tail()
```

```python
result = pd.concat([b_feat, df_authors_reduced], axis=1, sort=False, join='inner')
```
