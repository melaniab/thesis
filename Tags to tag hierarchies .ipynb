# Tags to hierarchies 
In this notebook, we will try to construct a reasonable term hierarchy from the tags using a mixed approach between co-occurences and density-based clustering. 

```python
import pandas as pd 
import numpy as np
```

```python
#load clusters data 
df_clusters = pd.read_csv("data/stags_clusters_1000.csv")
df_clusters.head()
```

```python
n_clusters = max(df_clusters["db_scan_bert"])
```

```python
for i in range(-1, 83): 
    print("cluster", i, ":", list(df_clusters[df_clusters["db_scan_bert"]==i]["tag_id"].values))
```

```python
bert_clusters = [] 
for i in range(0, n_clusters): 
    bert_clusters.append(list(df_clusters[df_clusters["db_scan_bert"]==i]["tag_id"].values))
    
    
```

```python
print(bert_clusters[2])
```

```python
df_bt_reduced = pd.read_csv("book_tags_reduced.csv", index_col = 0)
```

```python
df_bt_reduced
```

```python
df_occur = df_bt_reduced.groupby('tag_id').agg({'goodreads_book_id':lambda x: [_ for _ in x] }, 
                                   )
```

```python
df_occur["count_occur"] = df_occur["goodreads_book_id"].apply(len)
```

```python
df_occur["tag_id"] = df_occur.index
```

```python
df_occur.head(20)
```

```python
df_cl2 = df_occur.loc[bert_clusters[2]]
df_cl2.dropna(inplace=True)
df_cl2
```

```python
df_cl2["num_books"] = df_cl2["goodreads_book_id"].apply(len)
```

```python
df_cl2["num_books"].idxmax()
```

```python
df_cl2["goodreads_book_id"][1659]
```

```python
df_tags = pd.read_csv("tags.csv")
df_tags.loc[1659]["tag_name"]
```

```python
related_tags = [] 
for row in df_cl2.iterrows(): 
#     print(row[1])
    if set(row[1]["goodreads_book_id"]).issubset(set(df_cl2["goodreads_book_id"][1659])): 
        print(row[1]["tag_id"])
        related_tags.append(int(row[1]["tag_id"]))
```

```python
related_tags_names = [] 
for index in related_tags:
    related_tags_names.append(df_tags.loc[index]["tag_name"]) 
    print(df_tags.loc[index]["tag_name"])
```
